```
import pandas as pd
```

```
fullData = pd.ExcelFile('/content/gastos_limites.xlsx')
```

**Separando as abas em dataframes - gastos**

```
# 'extrai' aba gastos
df1 = pd.read_excel(fullData, sheet_name='gastos')
```

**Separando as abas em dataframes - limites**

```
# 'extrai' aba limites
df2 = pd.read_excel(fullData, sheet_name='limites')
```

```
df2.head(5)
```

**Retirando o símbolo de R$ da coluna valor_operação e transformando-o em float**

```
df1['valor_operacao'] = df1['valor_operacao'].str.lstrip('R$')
df1['valor_operacao'] = [x.replace(",", ".") for x in df1['valor_operacao']]
df1['valor_operacao'] = df1['valor_operacao'].astype(float)
df1['valor_operacao']
```

```
df1.dtypes
```

```
df1.head(5)
```

**Padronizando o nome da coluna nm_cliente do dataframe 1 (df1) para cliente**

```
df1 = df1.rename({'nm_cliente': 'cliente'}, axis = 1)
```

```
df1
```

**Retirando o símbolo de R$ da coluna limite e transformando-o em float**

```
df2['limite'] = df2['limite'].str.lstrip('R$')
df2['limite'] = [x.replace(",", ".") for x in df2['limite']]
df2['limite'] = df2['limite'].astype(float)
df2['limite']
```

```
df2.head(5)
```

**Unificando os dataframes df1 e df2 com base no nome do cliente**

```
df_new = df1.merge(df2, on = "cliente", how = "left")
```

```
df_new
```

**Verificando se todas as linhas tem correspondentes e nãão apresentan nem valores nulos tãão pouco NaN**

```
df_new.isna().sum()
```

```
df_new.isnull().sum()
```

```
df_new['dt_vencimento']
```

**Transformando as colunas dt_vencimento e dt_operação em datetime**

```
df_new['dt_vencimento'] = pd.to_datetime(df_new['dt_vencimento'])
df_new.dtypes
```

```
df_new['dt_operacao'] = pd.to_datetime(df_new['dt_operacao'])
df_new.dtypes
```

```
for index, row in df_new.iterrows():
  print(row['dt_vencimento'])
```

**Incluindo a coluna prox_mes para saber se o gasto foi feito antes ou apóós a data de vencimento**

```
for index, row in df_new.iterrows():
  if (row['dt_vencimento']< row['dt_operacao']):
    df_new.loc[[index],['prox_mes']] = 1
  else:
    df_new.loc[[index], ['prox_mes']]  = 0


df_new.head(20)
```

**Pelo valor associado à coluna prox_mes, determinou-se quando a fatura deveria ser paga:**

**- em janeiro, caso a data de operação tenha sido antes da data de vencimento;** 


**- fevereiro, caso a data de operação tenha sido após o vencimento**

```
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
for index, row in df_new.iterrows():
  if (row['prox_mes'] == 0):
    for i in range(0, len(meses), 1):
      if (i<= row['num_parcelas']):
        df_new.loc[[index],[meses[i]]] = row['valor_operacao']/row['num_parcelas']
      else:
        df_new.loc[[index], [meses[i]]] = 0

  else:
    df_new.loc[[index],[meses[0]]] = 0
    for i in range(1,len(meses), 1):
      if (i<= row['num_parcelas']):
        df_new.loc[[index],[meses[i]]] = row['valor_operacao']/row['num_parcelas']
      else:
        df_new.loc[[index], [meses[i]]] = 0    
        

    
print(df_new.iloc[:, 10:])
```

```
df_new.round(2)
```

**Agrupando por cliente e somando por mês suas despesas**

```
tabela_gastos = df_new[['cliente', 'janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']]
```

```
sum_gastos = tabela_gastos.groupby(['cliente']).sum()
sum_gastos.round(2)
```

**Incluindo a coluna limite na tabela_final**

```
tabela_final = sum_gastos.merge(df2, on = "cliente", how = "left")
tabela_final = tabela_final.drop(['Unnamed: 0', 'dt_vencimento'], axis=1)
```
